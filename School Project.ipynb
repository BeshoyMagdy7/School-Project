```python
import pandas as pd
schools = pd.read_csv("schools.csv")
schools.head()
```

```python
# twenty missing values in the percent_tested column.
schools.isna().sum() 
```

1. Finding schools with the best math scores

```python
# Finding schools with the best math scores (math scores of at least 80%)
best_math_schools = schools[schools['average_math']>=640].sort_values('average_math', ascending = False)
best_math_schools = best_math_schools[['school_name','average_math']]
best_math_schools
```

2. Identifying the top 10 performing schools

```python
# create new column (total_SAT) sum of three column(average_math , average_reading , average_writing).
schools["total_SAT"] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
schools
```

```python
# after add new column (total_SAT) sort data frame ascending by total_SAT
# Find the 10 best performing schools based on the total score across the three SAT sections.
top_10_schools = schools.sort_values("total_SAT",ascending=False)[['school_name','total_SAT']].head(10)
top_10_schools
```

3. Locating the NYC borough with the largest standard deviation in SAT performance

```python
# Locating the NYC borough with the largest standard deviation in SAT performance
NYC = schools.groupby('borough')['total_SAT'].agg(['count','mean','std'])
NYC
```

```python
# find largest_std_dev from NYC data frame
largest_std_dev = NYC[NYC['std'] == NYC['std'].max()]
largest_std_dev
```

```python
# rename columns of largest_std_dev data frame
largest_std_dev = NYC.rename(columns = {'count':'num_schools' , 'mean':'average_SAT' , 'std':'std_SAT'}).round(2)
largest_std_dev
```
